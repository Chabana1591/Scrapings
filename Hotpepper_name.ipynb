```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
from time import sleep
```

```python
def dellist(items,indexes):
    for index in sorted(indexes, reverse = True):
        del items[index]
```

```python
# import re
def GetLastPageIndex(bs,ul_selector='ul.pageLinkLinearBasic'):
    page_index_list = []
    for li in bs.select(ul_selector)[0].select('li'):
        match_tmp = re.search(r'[0-9]+',li.text)
        if match_tmp:
            page_index_list.append(int(match_tmp.group(0)))
    last_page_index = max(page_index_list)
    return last_page_index
```

```python
# import requests
page_index = 1
url = "https://www.hotpepper.jp/SA53/Y886/X550/G001_G012/bgn"+str(page_index)+"/"
response = requests.get(url)
print("response.status_code:"+str(response.status_code))
# from bs4 import BeautifulSoup
bs = BeautifulSoup(response.content,"lxml")
print("bs ok.")
last_page_index = GetLastPageIndex(bs,'ul.pageLinkLinearBasic')
last_page_index
```

```python
shop_info = {"name":[],"food_link":[],"drink_link":[]}
for page_index in range(1,last_page_index+1):
    print("page:"+str(page_index))
    # import requests
    url = "https://www.hotpepper.jp/SA53/Y886/X550/G001_G012/bgn"+str(page_index)+"/"
    response = requests.get(url)
    print("response.status_code:"+str(response.status_code))
    # from bs4 import BeautifulSoup
    bs = BeautifulSoup(response.content,"lxml")
    print("bs ok.")
    #         
    shop = bs.select('h3.shopDetailStoreName')
    del_index = []
    for i,s in enumerate(shop):
        if ("pr" in s.select('a')[0].get('onclick')):
            del_index.append(i)
    dellist(shop,del_index)
    print("del_index:"+str(del_index))
    #
    for s in shop:
        shop_info["name"].append(s.select('a')[0].text)
        shop_info["food_link"].append("https://www.hotpepper.jp"+s.select('a')[0].get("href")+"food/")
        shop_info["drink_link"].append("https://www.hotpepper.jp"+s.select('a')[0].get("href")+"drink/")
    print("sleep...")
    sleep(1)
    print("end")
print(shop_info)
```

```python
for i,name in enumerate(shop_info['name']):
#     name_tmp = re.sub(r'【.+】',"",name)
    name_tmp = name
    name_tmp = name_tmp.replace('\n',"")
    name_tmp = name_tmp.replace('\t',"")
    shop_info['name'][i]=name_tmp
```

```python
# import pandas as pd
shops_dt = pd.DataFrame(shop_info)
shops_dt
```

```python
shops_dt.to_csv('hotpepper_kokubuncho.csv')
print("successed.")
```
