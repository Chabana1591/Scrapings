```python
import requests
from bs4 import BeautifulSoup
import re
from time import sleep
import pprint
```

```python
# import requests
url = "https://ncode.syosetu.com/n6266ci/"
response = requests.get(url)
response.status_code
```

```python
from bs4 import BeautifulSoup
bs = BeautifulSoup(response.content,"lxml")
```

```python
bs.select('#novel_ex')[0].text
```

```python
m=bs.select('div.index_box dt.long_update')[0]
update=m.text.replace('\n','')
tmp = m.select('span')[0].get('title')
last_update = tmp if tmp else ""
update += last_update
update
```

```python
box = bs.select('div.index_box')[0].contents
box = list(filter(lambda x: x != '\n',box))
```

```python
chapter = []
chapter_index = []
subtitle = []
url = []
update = []
for i,elem in enumerate(box):
    if re.search('div',str(elem)):
        chapter.append(elem.text)
        chapter_index.append(((i+1)-len(chapter)))
    if elem.select('dd'):
        subtitle.append(elem.select('a')[0].text)
        url.append('https://ncode.syosetu.com'+elem.select('a')[0].get('href'))
        update_tmp = elem.select('dt.long_update')[0].text.replace('\n','')
        update_tmp_kai = elem.select('span')[0].get('title') if elem.select('span') else ""
        update.append(update_tmp + update_tmp_kai)
dictionary = {
    "chapter":chapter,
    "chapter_index":chapter_index,
    "subtitle":subtitle,
    "url":url,
    "update":update
}
```

```python
dic = dictionary
```

```python
dic['honbun']=[]
for u in dic['url']:
    # import requests
    print(u+"request")
    url = u
    response = requests.get(url)
    print(response.status_code)
    #from bs4 import BeautifulSoup
    bs = BeautifulSoup(response.content,"lxml")
    dic['honbun'].append(bs.select('p.novel_subtitle')[0].text+"\n\n")
    dic['honbun'][-1] += bs.select('#novel_honbun')[0].text
    print("sleepng...")
    sleep(5)
print("complete.")
```

```python
dic["chapter_index"]
```

```python
dic_samelen = {
    "subtitle":dic['subtitle'],
    "url":dic['url'],
    "update":dic['update'],
    "honbun":dic['honbun'],
    "chapter":[]
}
j = -1
for i,index in enumerate(dic["subtitle"]):
    if i in dic['chapter_index']:
        j += 1
    dic_samelen["chapter"].extend([dic['chapter'][j]])
print(len(dic_samelen['chapter']))
print(len(dic_samelen['subtitle']))
```

```python
import pandas as pd
shousetu_dt = pd.DataFrame.from_dict(dic_samelen)
shousetu_dt.head(3)
```

```python
shousetu_dt.to_csv("kusuriya.csv")
print('success')
```
