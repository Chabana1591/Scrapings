```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
from time import sleep
```

```python
urls_dt = pd.read_csv('Scraping/hotpepper_kokubuncho.csv')
urls_dt.head(3)
```

```python
for index in range(len(urls_dt)):
    print(urls_dt.at[index,'food_link'])
```

```python
shop = {'food':[],'drink':[]}
for index in range(20):
# for index in range(len(urls_dt)):
    print(index , urls_dt.at[index,'food_link'])
    # import requests
    url = urls_dt.at[index,'food_link']
    response = requests.get(url)
    print("response.status_code:"+str(response.status_code))

    # from bs4 import BeautifulSoup
    bs = BeautifulSoup(response.content,"lxml")

    shop['food'].append(bs.select('div.shopInner')[0])
    sleep(3)
    
    print(urls_dt.at[index,'drink_link'])
    # import requests
    url = urls_dt.at[index,'drink_link']
    response = requests.get(url)
    print("response.status_code:"+str(response.status_code))

    # from bs4 import BeautifulSoup
    bs = BeautifulSoup(response.content,"lxml")
    print("bs ok.")
    
    shop['drink'].append(bs.select('div.shopInner')[0])
    sleep(3)
    
    
print("finished.")
    
```

```python
shop_df = pd.DataFrame(shop)

df = pd.concat([urls_dt.head(20), shop_df], axis =1)
df
```

```python
json = {'json':[]}
for index in range(20):
    url = urls_dt.at[index,'food_link'].replace('food/','')
    print(index , url)
    # import requests
    response = requests.get(url)
    print("response.status_code:"+str(response.status_code))
    # from bs4 import BeautifulSoup
    bs = BeautifulSoup(response.content,"lxml")

    json['json'].append(bs.find(type="application/ld+json").get_text().replace('\n','').replace('\t',''))
    sleep(3)
json
```

```python
json_df = pd.DataFrame(json)

df = pd.concat([urls_dt.head(20), shop_df,json_df], axis =1)
df
```

```python
df.to_csv("Scraping/hotpepper.csv")
print("succeed")
```
